```python
import rosbag
from std_msgs.msg import String  # Import the message type you expect to read

# Specify the path to your ROS bag file
bag_path = '/home/annika/ITRL/GPS_frame/rosbag/2024-01-23-11-08-25.bag'
FixedFrameState1 = []
Lat1 = []
Long1 = []
# Open the bag file
with rosbag.Bag(bag_path, 'r') as bag:
    # Iterate through messages in the bag
    for topic, msg, t in bag.read_messages():
        # Process the messages based on the topic
        if topic == '/fixedgps/state':
            FixedFrameState1.append([msg.x,msg.y])
        if topic == '/gps/filtered':
            Lat1.append(msg.latitude)
            Long1.append(msg.longitude)
# Note: Replace '/your/topic' with the actual topic you are interested in
# You also need to import the correct message type for the specified topic
```

```python
import rosbag
from std_msgs.msg import String  # Import the message type you expect to read

# Specify the path to your ROS bag file
bag_path = '/home/annika/ITRL/GPS_frame/rosbag/2024-01-23-11-12-00.bag'
FixedFrameState2 = []
Lat2 = []
Long2 = []

# Open the bag file
with rosbag.Bag(bag_path, 'r') as bag:
    # Iterate through messages in the bag
    for topic, msg, t in bag.read_messages():
        # Process the messages based on the topic
        if topic == '/fixedgps/state':
            FixedFrameState2.append([msg.x,msg.y])
        if topic == '/gps/filtered':
            Lat2.append(msg.latitude)
            Long2.append(msg.longitude)
# Note: Replace '/your/topic' with the actual topic you are interested in
# You also need to import the correct message type for the specified topic
```

```python
import matplotlib.pyplot as plt

x_values1, y_values1 = zip(*FixedFrameState1)
x_values2, y_values2 = zip(*FixedFrameState2)

# Create a scatter plot for the first set of coordinates
plt.scatter(x_values1, y_values1, marker='o', label='Coordinates 1')

# Create a scatter plot for the second set of coordinates
plt.scatter(x_values2, y_values2, marker='s', label='Coordinates 2')

# Add labels and title
plt.xlabel('X-axis')
plt.xlim(10,20)
plt.ylim(5,20)
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Coordinates')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
